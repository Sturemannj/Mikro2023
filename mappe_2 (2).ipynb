```python
import pandas as pd
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
def process_traffic_data(file_path):
    # Les csv-filen
    traffic_data = pd.read_csv(file_path, sep=',', encoding= 'unicode_escape')
    # Fjern unødvendige kolonner
    columns_to_drop = ['Unnamed: 0','Fra','Til','Vegreferanse','Felt']
    traffic_data.drop(columns_to_drop, inplace=True, axis=1)
    # Pivoter dataframen
    traffic_data = traffic_data.pivot(index = 'Dato', columns= 'Navn', values = 'Trafikkmengde')
    # Fjern kolonneakser
    traffic_data = traffic_data.rename_axis(None, axis=1)
    # Reset indeksen
    traffic_data = traffic_data.reset_index()
    # Konverter 'Dato' kolonnen til datetime
    traffic_data['Dato']= pd.to_datetime(traffic_data['Dato'])
    return traffic_data
Trafikkmengde_etter_bompenger = process_traffic_data('Bompenger 3.csv')
```
