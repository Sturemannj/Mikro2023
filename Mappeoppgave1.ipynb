```python
import json
import pandas as pd
import requests
from pyjstat import pyjstat
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

postUrl = 'https://data.ssb.no/api/v0/no/table/10638/'
apiQuery = {
  "query": [
    {
      "code": "Reiseliv",
      "selection": {
        "filter": "item",
        "values": [
          "0",
          "01.01",
          "01.02",
          "01.03",
          "01.04",
          "01.05",
          "01.06",
          "01.08"
        ]
      }
    },
    {
      "code": "ContentsCode",
      "selection": {
        "filter": "item",
        "values": [
          "FastePriser"
        ]
      }
    },
    {
      "code": "Tid",
      "selection": {
        "filter": "item",
        "values": [
          "2007",
          "2008",
          "2009",
          "2010",
          "2011",
          "2012",
          "2013",
          "2014",
          "2015",
          "2016",
          "2017",
          "2018",
          "2019",
          "2020"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat2"
  }
}

def apiToDataframe(postUrl, query):
    res = requests.post(postUrl, json=query)
    ds = pyjstat.Dataset.read(res.text)
    df = ds.write('dataframe')
    df_id = ds.write('dataframe', naming='id')
    return df, df_id, ds

df_turistkonsum, df_id, ds = apiToDataframe(postUrl, apiQuery)
df_turistkonsum
```

- det vil føre til mere inntekt for lokale myndigheter: turistskatten vil gi økt inntekt for lokale myndigheter skal brukes for å forberede turistfasiliteter, infrastruktur og fremtidige prosjekter. 

- redusering av avhengig turistindustri: turistskatten vil føre til en mindre avhengig økonomi av turistindustrien: Betyr at vi vil ha en redusert avhengighet til turistindustrien som kan føre til mer bærekraftig vekst for å redusere sårbarheten for svingninger. 

- bevisste turister: turistskatten vil føre til at turister er mer bevisste på hvor de skal dra og hva den koster til de forskjellige stasjonene de skal reise, så med dette er veldig bukke mer bærekraftig og ansvarlig overnattingssteder. dette vil føre til mer bærekraftig, ansvarlig på miljøvennlige     attraksjoner.

- kvalitet på turistopplevelser: men dere skatt ville øke kvaliteten på turistopplevelser, dette betyr at kvaliteten på turistattraksjoner infrastruktur vil være økt, som gjør at tilgangen informasjon og tjenester for turister vil være bedre. 

### Kapittel 4 - Konklusjon
Basert på analysen som vi har konkludert med over med innføring av turistskatt på både positive og negative konsekvenser for en destinasjon. På den ene siden kan vi se at den generelle skatteinntekter kan brukes til å finansiere ulike tiltak for å øke bærekraftighet og forberede infrastruktur, men på en annen side kan vi se at en reduksjon i turisttrafikk som kan påvirke destinasjonens konkurranseevne spesielt hvis destinasjonen allerede er dyrere enn sine konkurrenter. slik at vi har sett under covid19 så vil en minske turists bølge fører til en nedgang i turist næringen men også grenene som er rettet mot transport spesielt fly og andre transportmidler ettersom at de blir hardest rammet av en minsket antall turister som reiser med fly. 

Effektene av en turistskatt til avhengig av ulike faktorer som størrelsen på skatten og hvordan inntektene forvandles og det stasjons prisnivå, dette vil også påvirke turistenes preferanser og etterspørsler. Derfor er det svært viktig at vi vurderer ekstremt grundig de ulike faktorene før man innfører en turistskatt. 

Men hvis det skal maksimere effekten og fordelingseffekten av en turistskatt bør skatte inntektene brukes til å finansiere bærekraftige tiltak og infrastruktur som vi øker kvaliteten på turistopplevelsen og destinasjonens at attraktivitet. Det er også derfor viktig å kommunisere tydelig med turistene om formålet med turistskatten og hvordan pengene blir brukt. 

Så mitt råd til nærings og fiskeridepartementet er å vurdere de ulike faktorene som kan forventes til effekten av en turistskatt å innføre en god ordning så vil ha en god oversikt over forvandling. Men også ha en god forståelse for lokale interesser slik at vi finner en passende løsning som tar hensyn til både økonomiske og miljømessige faktorer for fremtiden.
