```python
df = pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv",delimiter=";")

# Oppgave 7.1.A

# lager nye verdier for log p og log q  til p og q
df['P'] = np.exp(df['log p'])
df['Q'] = np.exp(df['log q'])
```

```python
# OPPGAVE 7.1.B
# Lager en figur for P
plt.plot(df['Year'], df['P'], marker='o', markersize=5, markerfacecolor='blue', markeredgecolor='black')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Price over Time')
plt.grid(True)
plt.xlim(df['Year'].min(), df['Year'].max())
plt.ylim(df['P'].min(), df['P'].max())
plt.legend(['Price'])
plt.style.use('ggplot')
plt.show()
```

```python
# OPPGAVE 7.1.B
# lager en figur for Q
plt.plot(df['Year'], df['Q'], marker='o', markersize=5, markerfacecolor='blue', markeredgecolor='black')
plt.xlabel('Year')
plt.ylabel('Crop Quantity')
plt.title('Crop Quantity over Time')
plt.grid(True)
plt.xlim(df['Year'].min(), df['Year'].max())
plt.ylim(df['Q'].min(), df['Q'].max())
plt.legend(['Crop Quantity'])
plt.style.use('ggplot')
plt.show()
```

```python
#Oppgave 7.1.2 A,B OG C

# lager liste for verdier
Q_values = list(range(20, 105, 5))

# lager dataframe med q verid 
df_table = pd.DataFrame({'Q': Q_values})

# legger til logQ linje
df_table['logQ'] = np.log(df_table['Q'])

# Legger til logP linje
df_table['logP_supply'] = -2 + 1.7*df_table['logQ']

# Legger til logP_Demand linje
df_table['logP_demand'] = 8.5 - 0.8*df_table['logQ']

# Legger til p_supply linje
df_table['P_supply'] = np.exp(df_table['logP_supply'])

# Legger til P_demand linje
df_table['P_demand'] = np.exp(df_table['logP_demand'])

# Print the table
print(df_table)
```

```python
#Oppgave 7.1.2 D
# plotter supply
plt.plot(df_table['Q'], df_table['P_supply'], label='Supply')

# plotter demandcurve
plt.plot(df_table['Q'], df_table['P_demand'], label='Demand')

plt.legend()

plt.xlabel('Quantity (millions)')
plt.ylabel('Price (thousands/watermelon)')

plt.show()
```

```python
#Oppgave 7.1.3 A og B

df_table['logP_new_supply'] = -2 + 1.7*df_table['logQ'] + 0.4
df_table['P_new_supply'] = np.exp(df_table['logP_new_supply'])

#lager verdier
Q_values = list(range(20, 105, 5))
#lager dataaframe med Q som verdi
df_table = pd.DataFrame({'Q': Q_values})
#setter logQ til linje 
df_table['logQ'] = np.log(df_table['Q'])
#setter logP til linje 
df_table['logP_supply'] = -2 + 1.7*df_table['logQ']
#setter logP_Demand til linje 
df_table['logP_demand'] = 8.5 - 0.8*df_table['logQ']
#setter p_supply til linjee
df_table['P_supply'] = np.exp(df_table['logP_supply'])
#setter P_demand til linje 
df_table['P_demand'] = np.exp(df_table['logP_demand'])
#setter logp_new:supply til linje
df_table['logP_new_supply'] = -2 + 1.7*df_table['logQ'] + 0.4
#setter P_new_supply til linje
df_table['P_new_supply'] = np.exp(df_table['logP_new_supply'])
#Plotter frafer
plt.plot(df_table['Q'], df_table['P_supply'], label='Supply')
plt.plot(df_table['Q'], df_table['P_demand'], label='Demand')
plt.plot(df_table['Q'], df_table['P_new_supply'], label='New Supply')
plt.xlabel('Quantity (millions)')
plt.ylabel('Price (dollars)')
plt.legend()
plt.show()
#Regner elasiteten
df_table['elasticity_supply'] = (df_table['logQ'] - 2) / (df_table['logP_supply'] + 2)
df_table['elasticity_new_supply'] = (df_table['logQ'] - 2) / (df_table['logP_new_supply'] + 2)

print(df_table[['Q','elasticity_supply','elasticity_new_supply']])
```
